```
#%cd /content/img/img_train
%cd /content/drive/My\ Drive/GWC_imgs/dataset/train/
c = 1
dat = []
for img in iag_lst:
    image = imageio.imread(img) 
    label = img
    bbx = []
    for b in df.itertuples():
      if b[1]==img:
        bbx.append(list(b[1:]))
    bbs = BoundingBoxesOnImage([BoundingBox(bb[4],bb[5],bb[6],bb[7]) for bb in bbx],shape=image.shape)


    image_aug, bbs_aug = iaa.Affine(rotate=10,fit_output=True)(image=image, bounding_boxes=bbs)
    img_aug = ia.imresize_single_image(image_aug,(1024,1024))
    bbx_aug = bbs_aug.to_xyxy_array()
    bbx_aug = bbx_aug*image.shape[0]/image_aug.shape[0]
    imageio.imwrite(label.replace(".jpg","rot.jpg"),img_aug)
    d = [label.replace(".jpg","rot.jpg"),bbx_aug]
    dat.append(d)
    #bbs_aug = BoundingBoxesOnImage([BoundingBox(bb[0],bb[1],bb[2],bb[3]) for bb in bbx_aug],shape=image.shape)
    #ia.imshow(bbs_aug.draw_on_image(img_aug,size = 5))


    image_aug, bbs_aug = iaa.weather.Fog()(image=image, bounding_boxes=bbs)
    bbx_aug = bbs_aug.to_xyxy_array()
    imageio.imwrite(label.replace(".jpg","fog.jpg"),image_aug)
    d = [label.replace(".jpg","fog.jpg"),bbx_aug]
    dat.append(d)


    image_aug, bbs_aug = iaa.contrast.AllChannelsCLAHE()(image=image, bounding_boxes=bbs)
    imageio.imwrite(label.replace(".jpg","cont.jpg"),image_aug)
    d = [label.replace(".jpg","cont.jpg"),bbx_aug]
    dat.append(d)


    image_aug, bbs_aug = iaa.weather.Clouds()(image=image, bounding_boxes=bbs)
    imageio.imwrite(label.replace(".jpg","cld.jpg"),image_aug)
    d = [label.replace(".jpg","cld.jpg"),bbx_aug]
    dat.append(d)
    if c%100 == 0:
      print(c)
    c = c + 1
```

```
final = []
for d in dat:
  k = []
  k.append(d[0])
  k.append(1024)
  k.append(1024)
  k.append("wheat")
  # k.append(0)
  # k.append(0)
  # k.append(0)
  # k.append(0)
  for i in d[1]:
    k = []
    k.append(d[0])
    k.append(1024)
    k.append(1024)
    k.append("wheat")
    k.append(i[0])
    k.append(i[1])
    k.append(i[2])
    k.append(i[3])
    final.append(k)
  #print(k)
print("done")
```
